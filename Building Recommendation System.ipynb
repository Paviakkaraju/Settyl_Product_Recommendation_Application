# Normalization

```python
from sklearn.preprocessing import MinMaxScaler

m = MinMaxScaler()
df['price_num_log'] = m.fit_transform(df[['price_num_log']])
df['rating_count_log'] = m.fit_transform(df[['rating_count_log']])
```

# Building a Recommendation System

```python
df.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
features = ['rating', 'rating_count_log', 'price_num_log']
```

```python
# Suggesting recommendations based on category and sub-category
# Lets assume that this is the USER INPUT
cate = 'Computers&Accessories'
sub_cat = 'Cables&Accessories'
```

```python
display_df = display[(display['main_category']==cate)&(display['sub-category']==sub_cat)]
```

```python
filter_df = df[(df['main_category']==cate)&(df['sub-category']==sub_cat)]
```

```python
filter_df['combined']  = filter_df['product_name']+" "+filter_df['about_product']
vect = TfidfVectorizer(stop_words='english', max_df=0.95, min_df=2, ngram_range=(1, 1))
tfidf_matrix = vect.fit_transform(filter_df['combined'])
```

```python
df_num = filter_df[features]
tfidf_matrix.shape, df_num.shape
```

```python
df_num.isna().sum()
```

```python
scaled_numerical_matrix = np.array(df_num)
print(np.isnan(scaled_numerical_matrix).sum())
```

```python
from scipy.sparse import hstack
combined_matrix = hstack([tfidf_matrix, df_num])
```

```python
cosine_sim = cosine_similarity(combined_matrix, combined_matrix)
```

# Recommending 

```python
recommendations = {}
product_names = filter_df['product_name'].tolist()

for idx, row in enumerate(cosine_sim):
    similar_products = np.argsort(row)[::-1][1:6]
    similar_product_names = filter_df.iloc[similar_products]['product_name'].tolist()
    suggestions_ind = filter_df.iloc[similar_products]['product_name'].index.tolist()
```

```python
for i in suggestions_ind:
    print(display_df.loc[i])
    print()
```
