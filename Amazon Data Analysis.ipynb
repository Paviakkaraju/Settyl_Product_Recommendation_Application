# About the Dataset

Source: https://www.kaggle.com/datasets/karkavelrajaj/amazon-sales-dataset/data

This dataset is having the data of 1K+ Amazon Product's Ratings and Reviews as per their details listed on the official website of Amazon

**Features**

* **product_id** - Product ID
* **product_name** - Name of the Product
* **category** - Category of the Product
* **discounted_price** - Discounted Price of the Product
* **actual_price** - Actual Price of the Product
* **discount_percentage** - Percentage of Discount for the Product
* **rating** - Rating of the Product
* **rating_count** - Number of people who voted for the Amazon rating
* **about_product** - Description about the Product
* **user_id** - ID of the user who wrote review for the Product
* **user_name** - Name of the user who wrote review for the Product
* **review_id** - ID of the user review
* **review_title** - Short review
* **review_content** - Long review
* **img_link** - Image Link of the Product
* **product_link** - Official Website Link of the Product

# Product and Price Analysis

```python
df[price_col].describe()
```

```python
df['price_num'] = df['actual_price'].str.replace(',', '').str.replace('₹', '').astype('float32')
```

```python
df['disc_price_num'] = df['discounted_price'].str.replace(',', '').str.replace('₹', '').astype('float32')
```

```python
df['discount_percent_num'] = df['discount_percentage'].str.replace('%', '').astype('int32')
```

```python
df.info()
```

```python
num = ['price_num', 'disc_price_num', 'discount_percent_num']
summary = df[num].describe().T.reset_index()
summary['skewness'] = summary['index'].map(lambda x: round(df[x].skew(), 3))
summary['kurtosis'] = summary['index'].map(lambda x: round(df[x].kurt(), 3))
summary
```

* The minimum actual price of a product is 39 rupees and the maximum price is 139900. The middle value is 1700.
* Discount percentage range from 0 to 94%

```python
# Category wise price and discount

# Top 10 products with highest discount
df.groupby('category')[['discount_percent_num']].median().sort_values(by='discount_percent_num',ascending=False).head(10)
```

```python
# Top 10 products with no discount
```

```python
df.groupby('category')[['discount_percent_num']].median().sort_values(by='discount_percent_num',ascending=False).tail(10)
```

```python
# Top 10 costly product categories
df.groupby('category')[['price_num']].median().sort_values(by='price_num',ascending=False).head(10)
```
